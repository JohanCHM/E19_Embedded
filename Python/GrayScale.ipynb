```python
struct ColorRGB8
{
    char r;
    char g;
    char b;

    ColorRGB8 toGrayScale()
    {
        ColorRGB8 ret;

        //Intensity Algorithm as per void_ptr's request
        char linearIntensity = (char)(0.2126f * r + 0.7512f * g + 0.0722 * b);
        ret.r = linearIntensity;
        ret.g = linearIntensity;
        ret.b = linearIntensity;
        return ret;
    }
};
```
